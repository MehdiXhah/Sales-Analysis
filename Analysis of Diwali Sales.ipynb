```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn

Frame = pd.read_csv("PRP1.csv")
Frame.head(10)
```

```python
Frame.info()
```

```python
Frame.describe()
```

```python
Frame.drop(["Staus"],axis = 1, inplace = True)
Frame
```

```python
Frame.isnull().sum()
```

```python
Frame["Marital"] = Frame["Marital"].fillna(2)
Frame
```

```python
Frame ["Occupatio"] = Frame["Occupatio"].fillna("others")
Frame
```

```python
Frame["Product_Orders"] = Frame["Product_Orders"].fillna(0.0)
Frame
```

```python
Frame["Amoun"]= Frame["Amoun"].fillna(0.0)
Frame
```

```python
Frame.columns
```

```python
#TO RENAME A COLUMN
Frame.rename(columns= {"Amoun":"Total Amount"})        
```

```python
Frame.rename(columns={"Occupatio":"Occupation"})
```

_**Exploratory Data Analysis**_


```python
ax = sn.countplot(data = Frame ,x = "Gender" )
ax.bar_label(ax.containers[0])
sn.set(rc = {"figure.figsize":(5,5)})
plt.show()
```

**From the graph I analysed that "Women" count was higher in purchasing the products.**

```python
Grouped_data = Frame.groupby("Gender")["Amoun"].sum()
print (Grouped_data)
#data = pd.DataFrame(Grouped_data)
colors = ["yellow","green","blue"]
a = sn.barplot(data = Frame,x = "Gender" , y = "Amoun", hue = "Marital" ,palette = color, dodge = True, errorbar = None )
sn.set(rc = {"figure.figsize":(8,6)})
#a.bar_label(a.containers[0])
for bars in a.containers:
    a.bar_label(bars)
plt.show()
```

**Here I analysed the most revenue generated from the different Genders based on their Marital Status and I found that most the revenue was generated
by the "Married Females".**

```python
b = sn.countplot(data = Frame, x = "Age Group", hue = "Gender", palette = "viridis")
for bars in b.containers:
    b.bar_label(bars)
sn.set(rc = {"figure.figsize":(5,5)})
plt.show
```

**By finding out from the above graph , I came to point that Females between the Age Group "26-35" did more shopping during dewali season.**

```python
Grouped_data = Frame.groupby(["Age Group"], as_index = False)["Amoun"].sum().sort_values(by = "Amoun", ascending = False)
print (Grouped_data)
colors = ["yellow","green","blue","orange","purple","red","brown"]
c = sn.barplot(data = Frame,x = "Age Group" , y = "Amoun",hue = "Age Group", errorbar = None, palette = colors, legend = False)
sn.set(rc = {"figure.figsize":(6,4)})
for bars in c.containers:
    c.bar_label(bars)
plt.xticks(rotation = 45)    
plt.tight_layout()
plt.show()
```

**Higher Revenue was generated by the people between age group of "51-55".**

```python
state_sale = Frame.groupby(["State"],as_index =False)["Product_Orders"].sum().sort_values(by= "Product_Orders", ascending = False)
sn.set(rc = {"figure.figsize":(9,5)})
sn.barplot(data = state_sale, x = "State", y = "Product_Orders", hue = "State", legend = False)
plt.xticks(rotation = 45)    
plt.tight_layout()
plt.show()
```

**Most of the clients ordered their products from the state "Karnatak".**

```python
state_sale = Frame.groupby(["State"],as_index =False)["Amoun"].sum().sort_values(by= "Amoun", ascending = False)
sn.set(rc = {"figure.figsize":(9,5)})
sn.barplot(data = state_sale, x = "State", y = "Amoun")#, hue = "State",legend = False)
plt.xticks(rotation = 45)    
plt.tight_layout()
plt.show()
```

**It was  analysed that Higher amount of revenue was generated aslo from the state "Karnatak".**
